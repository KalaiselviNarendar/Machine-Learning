```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('voice-classification.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
print('Shape : ',df.shape)
print(df.shape[0])
print("total no of label:{}".format(df.shape[0]))
print('total no of male:{}'.format(df[df.label=='male'].shape[0]))
print('total no of female:{}'.format(df[df.label=='female'].shape[0]))
```

```python
x=df.iloc[:,:-1]
x.head(2)
```

```python
print(df.shape)
print(x.shape)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
y=df.iloc[:,-1]
gender_encoder = LabelEncoder()
y=gender_encoder.fit_transform(y)
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x)
x=scaler.transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.metrics import classification_report,confusion_matrix
```

```python
svc_model=SVC()
svc_model.fit(x_train,y_train)
y_pred=svc_model.predict(x_test)
print('Acuracy score:',metrics.accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
# optimizing the result
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'C':[0.1,1,10,100],'gamma':[1,0.1,0.01,0.001]}
```

```python
grid=GridSearchCV(SVC(),param_grid,refit=True,verbose=2)
grid.fit(x_train,y_train)
```

```python
grid_predictions = grid.predict(x_test)
```

```python
print('Accuracy score:',metrics.accuracy_score(y_test,grid_predictions))
print(confusion_matrix(y_test,grid_predictions))
print(classification_report(y_test,grid_predictions))
```
